```python
import pandas as pd
```

```python
#Importing song spotify_dfset for EDA and classification 
spotify_df = pd.read_csv('spotify_songs.csv')
```

```python
#Checking for null values
spotify_df.isna().sum()
```

```python
spotify_df.head()
```

```python
spotify_df = spotify_df[spotify_df['language'] == 'en']
spotify_df.drop(columns=['language', 'playlist_name', 'playlist_id'], inplace=True)
spotify_df = spotify_df.drop_duplicates(subset=['track_name', 'track_artist'])
spotify_df['track_album_release_date'] = pd.to_datetime(spotify_df['track_album_release_date'], infer_datetime_format=True)
spotify_df = spotify_df.sort_values(by=['track_album_release_date'])
spotify_df.reset_index(drop=True, inplace=True)
```

```python
spotify_df.head()
```

```python
print(spotify_df.dtypes)
```

```python
#Creating a new spotify dataframe which only requires the necessary parameters for the classification
mod_spotify_df = spotify_df.drop(spotify_df.columns[0:10],axis = 1)
```

```python
#Checking the result of the modified spotify_dfset
mod_spotify_df.head()
```

```python
# Assuming 'mod_spotify_spotify_df' contains your standardized spotify_df
standardized_features = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms']

# Calculate mean and standard deviation for each feature
mean_values = mod_spotify_df[standardized_features].mean()
std_deviation = mod_spotify_df[standardized_features].std()

# Display the mean and standard deviation for each feature
for feature, mean, std in zip(standardized_features, mean_values, std_deviation):
    print(f"Feature: {feature}")
    print(f"Mean: {mean}")
    print(f"Standard Deviation: {std}")
    print()
```
